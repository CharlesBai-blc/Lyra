```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
dir = Path(".")
df1 = pd.read_csv(dir / "songs_condensed.csv")

plt.figure()

bars = np.zeros(20)
for i in range(8000):
    index = int(df1.danceability[i]//0.05)
    bars[index] = bars[index] + 1

for i in range(20):
    plt.bar(i*0.05, bars[i])
plt.show()
```

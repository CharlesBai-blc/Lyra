```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
dir = Path(".")
df1 = pd.read_csv(dir / "songs_condensed.csv")

plt.figure(1)
xvalues = np.arange(20)
bars = np.zeros(20)
for i in range(8000):
    index = int(df1.danceability[i]//0.05)
    bars[index] = bars[index] + 1
plt.bar(xvalues, bars, .9)
plt.show()


plt.figure(2)
plt.scatter(df1.danceability[:8000], df1.valence[:8000])
plt.xlabel('Danceability')
plt.ylabel('Valence')
plt.title('Danceability vs Valence')
plt.show()
```
